```python
# Import sample data

data = pickle.load(file=open('plot_data1.p'))
```

```python
data.keys()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(8,6)

plt.plot(data['train'])
plt.xticks(rotation=45, fontsize=16)
plt.yticks(fontsize=16)
plt.title('Training data', fontsize=24)
plt.xlabel('Timestamp', fontsize=16)
plt.ylabel('Price', fontsize=18)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(8,6)

plt.plot(data['test'])
plt.xticks(rotation=45, fontsize=16)
plt.yticks(fontsize=16)
plt.title('Test data', fontsize=24)
plt.xlabel('Timestamp', fontsize=16)
plt.ylabel('Price', fontsize=18)
```

```python
data['buy'].shape
```

```python

fig, ax = plt.subplots()
fig.set_size_inches(14,10)

y = np.cumsum(data['buy'][:,1,0])
x = data['buy'][:,0,0]
plt.plot(x, y)
ax.fill_between(x,y)

y = np.cumsum(data['sell'][:,1,0])
x = data['sell'][:,0,0]
plt.plot(data['sell'][:,0,0], y)
ax.fill_between(x,y)

plt.xticks(rotation=45, fontsize=16)
plt.yticks(fontsize=16)

plt.title('Bid/Ask book', fontsize=24)
plt.xlabel('Price (BTC)', fontsize=16)
plt.ylabel('Quantity', fontsize=18)

plt.show()
```
